# Using the Brandwatch API!



Lets first get a better idea about what an API is.

API stands for "Application Programming Interface". Think of it like an agreement between two people. 

"If you give me a set of instructions, I will perform this action or return this information"

# What is an SDK?!

An SDK, or "Software Development Kit" is something built inconjunction with an API. It's a set of tools that you use to build things for a specific API

# So what is the difference between the two? Do I always need an SDK?

An API's primary function is to integrate different pieces of software together. Like my laptop's operating system and the Brandwatch platform. It defines very specific questions that can be asked and very specifically formatted answers that will be provided.

Do you always need an SDK to worth with an API? Absolutely not! If an API is properly documents (ha.) then there is no need for an SDK. However, the common case is that if there is an SDK, it makes the process of writing code much, much easier

# A bit more terminology:


<li> Endpoints: think of this like an address .. a web address! </li>
<li> Calls: I use this to reference one request to an endpoint </li>
<li> Parameters: These are additional specifications we send. Like "queryId" or "startDate" or "endDate" </li>
<li> Rate Limits: These govern fair-usage policies. Rate limits don't affect how fast calls complete, rather how many calls you can make in a certain timeframe </li>
<li> HTTP Verbs: This denotes special web actions. The verb we will focus on today is GET. The other HTTP verbs are POST, PATCH, PUT, DELETE.
